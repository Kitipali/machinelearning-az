**Regresión Lineal Simple**


## Importamos bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importamos el dataset

```python
dataset = pd.read_csv('Salary_Data.csv') # está en la misma carpeta que este script
X = dataset.iloc[:, :-1].values # todas las columnas excepto la última
y = dataset.iloc[:, -1].values # la última columna
```

```python
dataset.head()
```

## Dividimos el dataset en entrenamiento y testing

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
```

## Crear modelo de Regresión Lineal Simple con el conjunto de entrenamiento

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(X_train, y_train)
```

## Predecir el conjunto de test

```python
y_pred = regression.predict(X_test) # ponemos la variable independiente de Test (X_test)
```

```python
y_pred
```

## Visualizar los resultados de entrenamiento

```python
y_test
```

```python
plt.scatter(X_train, y_train, color = 'red') # gráfico de dispersión
plt.plot(X_train, regression.predict(X_train), color = 'blue')
plt.title('Sueldo vs Años de Experiencia (Conjunto de entrenamiento)')
plt.xlabel('Años de Experiencia')
plt.ylabel('Sueldo en $')
plt.show()
```

## Visualiizar los resultados de Test

```python
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train, regression.predict(X_train), color = 'blue')
plt.title('Sueldo vs Años de Expeirencia (Conjunto de Testing)')
plt.xlabel('Años de experiencia')
plt.ylabel('Sueldo en $')
plt.show()
```
